```python
file_path = "./부산광역시_해수욕장 방문 현황_20221231.csv"
df1 = pd.read_csv(file_path,encoding="euc=kr")
plt.figure(figsize=(16,8))
sns.lineplot(x="연도별",y="방문자수(천명)",hue="해수욕장명",data=df1)
plt.xticks(rotation=45)
plt.title("부산광역시 해수욕장의 연도별 방문자수(천명)")
plt.show()
```

```python
### 파일 읽어들이기
file_path = "./연안별_수질등급_최종.xlsx"
df1 = pd.read_excel(file_path)

### pivot_table 생성
df1_pivot = df1.pivot_table(values="정점",
                index="등급",
                columns="해역",
               aggfunc="count")

###히트맵
plt.figure(figsize=(12,8))
sns.heatmap(df1_pivot,annot=True,fmt=".0f",cmap="YlGnBu")
plt.title("해역별 등급 현황")
plt.show()
```
