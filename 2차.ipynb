```python
### 파일 읽어들이기
file_path = "./해수욕장_기준코드_최종.xlsx"
df1 = pd.read_excel(file_path)
wea_df = pd.DataFrame()

### 날짜데이터 생성
from datetime import datetime

# 시작 날짜
start_date = datetime(2022, 7, 1)
# 종료 날짜
end_date = datetime(2022, 8, 31)

date_list = []
current_date = start_date

while current_date <= end_date:
    year = current_date.year
    month = current_date.month
    day = current_date.day

    # 년월일 형식으로 변환
    date_string = f"{year:04d}{month:02d}{day:02d}1600"  # 시간 정보 추가
    date_list.append(date_string)

    # 다음 날짜로 이동
    # 현재 날짜의 일(day)이 31일이면 다음 달로 이동
    if current_date.day == 31:
        # 현재 날짜의 월(month)이 12월이면 다음 년도로 이동
        if current_date.month == 12:
            current_date = current_date.replace(year=current_date.year + 1, month=1, day=1)
        # 다음 달로 이동
        else:
            current_date = current_date.replace(month=current_date.month + 1, day=1)
    # 현재 달에서 다음 날짜로 이동
    else:
        current_date = current_date.replace(day=current_date.day + 1)
        
### API호출하여 데이터 추출
for ii in range(0,len(date_list),1):
    for i in range(0,len(df1["순번"]),1):
        try:
            url=f"https://apis.data.go.kr/1360000/BeachInfoservice/getTwBuoyBeach?serviceKey=VZCe8aRGb%2FkA7AaLIcoYizusWQuR%2BxOv9HGc7ZEYIa6jCxP6NebL97arXaBBLhsTRX6P5cK4bFvOAObN3Ispaw%3D%3D&numOfRows=10&pageNo=1&dataType=XML&beach_num={df1['순번'][i]}&searchTime={date_list[ii]}"
            rs_data = requests.get(url)
            rs_text = rs_data.text
            rs_xml = BeautifulSoup(rs_text, "xml")
            wea_df1 = pd.DataFrame()
            wea_df1.loc[i, "해수욕장이름"]=df1["해수욕장이름"][i]
            wea_df1.loc[i, "날짜"]=rs_xml.find("tm").text
            wea_df1.loc[i, "수온"]=rs_xml.find("tw").text
            wea_df = pd.concat([wea_df,wea_df1],ignore_index=True)
        except AttributeError:
            wea_df1.loc[i, "날짜"] = ""
            wea_df1.loc[i, "수온"] = ""
            print("속성을 찾을 수 없습니다.")
            wea_df = pd.concat([wea_df, wea_df1], ignore_index=True)
wea_df
```
